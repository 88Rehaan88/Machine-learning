Checking for Outliers

```python
# Assuming daily_sales is your DataFrame with the 'Total' column having daily total sales.
mean_sales = daily_sales['Total'].mean()
std_sales = daily_sales['Total'].std()

# Define a threshold for outliers, commonly set to 2 or 3 standard deviations from the mean
threshold = 3

# Detect outl0iers which are beyond the threshold of standard deviations from the mean
outliers = daily_sales[(daily_sales['Total'] > mean_sales + 2 * std_sales) |
                       (daily_sales['Total'] < mean_sales - 2 * std_sales)]
outliers
```

### *Checking for Stationarity:*

```python
from statsmodels.tsa.stattools import adfuller
```

```python
# Perform the Augmented Dickey-Fuller test
adf_result = adfuller(daily_sales['Total'])
```

### Hypothesis test :
 Ho (Null Hypothesis) : Daily sales data is not stationary.

 H1 (Alternatice hypothesis) : Daily sales data is stationary


---> If the p-value is less than 0.05 and the ADF statistic is lower than the
     critical values, you can reject the null hypothesis and conclude that the series is stationary.


```python
print('ADF Statistic: %f' % adf_result[0])
print('p-value: %f' % adf_result[1])
print('Critical Values:')
for key, value in adf_result[4].items():
    print('\t%s: %.3f' % (key, value))

# A p-value less than 0.05 and an ADF statistic lower than the critical values at 1%, 5%, or 10% levels indicates stationarity.
```

***Conclusion***

1) ADF statistic : Since the ADF statistic(-7.654896) is much lower than the critical values at 1%, 5%, 10%, this suggests strong evidence against the null hypothesis.





2) p-value: A p-value of essentially 0 (0.000000) indicates that the null hypothesis of the ADF test (which posits that the series is non-stationary) can be rejected with high confidence.

Based on these results, we can conclude that the Total daily sales data is stationary. This means that the mean and variance of the series are constant over time, and the autocorrelation structure does not change.

```python
results
```

```python
# Summary of the model
print(results.summary())
```

```python
# Forecast
forecast = results.get_forecast(steps=len(test_data))
mean_forecast = forecast.predicted_mean
conf_int = forecast.conf_int()

# Plot the forecast along with the historical data
plt.figure(figsize=(10, 6))
plt.plot(train_data, label='Training Data')
plt.plot(test_data, label='Testing Data', color='orange')
plt.plot(mean_forecast.index, mean_forecast, label='Forecast', color='red')
plt.fill_between(conf_int.index, conf_int.iloc[:, 0], conf_int.iloc[:, 1], color='red', alpha=0.3)
plt.title('SARIMA Forecast')
plt.xlabel('Date')
plt.ylabel('Total Sales')
plt.legend()
plt.show()
```

```python
# Assuming that mean_forecast is a DataFrame and we need to extract the forecast column
# The code below assumes the name of the column containing the forecast values is 'forecast'
# Replace 'forecast' with the actual column name that holds your forecasted values

if isinstance(mean_forecast, pd.DataFrame):
    mean_forecast = mean_forecast['forecast']  # Replace 'forecast' with your actual forecast column name

if isinstance(test_data, pd.DataFrame):
    test_data = test_data['Total']  # Replace 'Total' with your actual test data column name

# Ensure you have no NaN values
valid_mask = ~test_data.isna() & ~mean_forecast.isna()

# Now calculate MAPE only on non-NaN values
mape = np.mean(np.abs((test_data[valid_mask] - mean_forecast[valid_mask]) / test_data[valid_mask])) * 100

print(f'The MAPE is: {mape}%')
```

```python
# Now you can safely calculate the MAPE
# Ensure you drop NaN values from both Series to avoid errors in the calculation
valid_mask = ~test_data.isna() & ~mean_forecast.isna()
mape = np.mean(np.abs((test_data[valid_mask] - mean_forecast[valid_mask]) / test_data[valid_mask])) * 100

print(f'The MAPE is: {mape}%')
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error
import numpy as np

# Calculate the errors
mae = mean_absolute_error(test_data, mean_forecast)
mse = mean_squared_error(test_data, mean_forecast)
rmse = np.sqrt(mse)  # or mse**(0.5)

# Print the errors
print(f'Mean Absolute Error (MAE): {mae}')
print(f'Mean Squared Error (MSE): {mse}')
print(f'Root Mean Squared Error (RMSE): {rmse}')
```
